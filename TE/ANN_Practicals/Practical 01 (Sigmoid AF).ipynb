```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(temp_arr):
    return 1 / (1 + np.exp(-temp_arr))
```

```python
arr = np.linspace(-8, 8)
plt.plot(arr, sigmoid(arr))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')

# only for grid
plt.grid()
plt.show()
print("\nName : Shantanu Anant Gaikwad\nRoll No : 27")
```
